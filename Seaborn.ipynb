```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("crop_production.csv")
```

```python
sns.scatterplot(data= a,x="Crop_Year",y="Production",hue="Production")
```

The scatterplot function in plotting the points in the graph. By plotting the points in the graph it represents how many crop production(y-axis) rate has increased in the years(x-axis),

```python
sns.relplot(data= a,x="Crop_Year",y="Production",kind="line")
```

We can access many types of plot using relational plots.In this we have used line plot to demonstrate how production rate has increased in the years.It is shows upside and down direction of the line that represents the increase of he production rate.

```python
sns.relplot(data=a,x="Crop_Year",y="Production", hue="Crop", height=10)
```

In this graph, the colours representing the types of crops used in the crop production. The production of these crops are used is represented as y-axis and how many productions has happened in a year is represented as x-axis.The colour representation is done by 'hue' parameter.

```python
sns.catplot(data=a, x="Production", y="Crop", kind="box",height=15)
```

A box plot shows the distribution of quantitative data in a way that facilitates comparisons between the crops and the crop production rate.

```python
sns.barplot(data=a, y="Production", x="Crop_Year")
```

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle between the production and the year
